```python
#doing our analysis
ncov_record = SeqIO.read("Downloads/sequence.fasta","fasta")
```

```python
ncov_record
```

```python
ncov_dna = ncov_record.seq
```

```python
#covid dna
ncov_dna
```

```python
len(ncov_dna)
```

```python
#protein synthesis
#DNA >RNA> Protein/AA
```

```python
#Transcription (DNA > mRNA)
ncov_mrna = ncov_dna.transcribe()
```

```python
ncov_mrna
```

```python
len(ncov_mrna)
```

```python
#translate to aminoacids/Protein
ncov_protein = ncov_mrna.translate()
```

```python
ncov_protein
```

```python
len(ncov_protein)
```

```python
#codon = 3
len(ncov_dna)/3
```

```python
#find all aminoacids before stop codon
ncov_protein
```

```python
ncov_aa = ncov_protein.split("*")
```

```python
ncov_aa
```

```python
ncov_clean = [ str(i) for i in ncov_aa]
```

```python
ncov_clean
```
